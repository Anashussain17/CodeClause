```python
import cv2
from fer import FER

# Initialize the emotion detector
emotion_detector = FER()

# Capture video from the webcam
video_capture = cv2.VideoCapture(0)

while True:
    # Read a frame from the webcam
    ret, frame = video_capture.read()
    
    if not ret:
        break

    # Detect emotions in the frame
    emotions = emotion_detector.detect_emotions(frame)

    # Display the detected emotions
    for emotion in emotions:
        x, y, w, h = emotion["box"]
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        emotion_text = max(emotion["emotions"], key=emotion["emotions"].get)
        cv2.putText(frame, emotion_text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)

    # Show the frame with detected emotions
    cv2.imshow("Facial Emotion Detection", frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close windows
video_capture.release()
cv2.destroyAllWindows()
```
